```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the barchart
labels = ['First', 'Second', 'Last', 'Next', 'Previous', 'Marked (j-th)']
in_context = [95, 95, 87.5, 96, 79, 97]
id_guided = [97.5, 5, 0, 81, 18.5, 34.5]
id_guided_recitation = [95, 87.5, 85, 85, 81, 98.5]

x = np.arange(len(labels))  # the label locations
width=0.2
# Define new colors for distinction
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

# Further increasing the text sizes for optimal readability

fig, ax = plt.subplots(figsize=(14, 6.5))  # Increased figure size for better text readability
rects1 = ax.bar(x - width, in_context, width, label='In-context', color=colors[0])
rects2 = ax.bar(x, id_guided, width, label='ID-guided', color=colors[1])
rects3 = ax.bar(x + width, id_guided_recitation, width, label='ID-guided + Recitation', color=colors[2])

# Further increase labels and title size
ax.set_xlabel('Experiment Setup', fontsize=18)
ax.set_ylabel('EM score', fontsize=18)
ax.set_title('Effect of Recitation', fontsize=20)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=16)

# Keeping the legend in the best location and increasing font size further
ax.legend(fontsize=16, loc='best', bbox_to_anchor=(1, 1))

# Increase the padding and fontsize for bar labels for even better readability
# ax.bar_label(rects1, padding=3, fontsize=14)
# ax.bar_label(rects2, padding=3, fontsize=14)
# ax.bar_label(rects3, padding=3, fontsize=14)

fig.tight_layout()

plt.show()
# write to pdf
fig.savefig('/home/aiops/zhuty/tinyllama/qe_prep/recitation_effect.pdf', bbox_inches='tight')
```
