```python
import torch
torch.manual_seed(0)
attn_weights = torch.randn(1, 32, 652, 652)
start_idx = 40
end_idx = 615
mask = torch.zeros((attn_weights.size()[2], attn_weights.size()[2]), dtype=torch.bool)
mask
square_size = end_idx - start_idx + 1
mask[-1,start_idx:start_idx + square_size] = True
mask[-1]
sub_matrix = attn_weights[:, :, -1, start_idx: start_idx + square_size]
mask = sub_matrix > 0
sub_matrix.masked_fill_(mask, float('-inf'))
attn_weights[:, :, -1, start_idx: start_idx + square_size] = sub_matrix
print(attn_weights[:, :, -1, :][0][0])
```
