```python
language_dist_dict = {}
for language in ['ind', 'vie','thai', 'en']:
    with open(f'/home/aiops/zhuty/tinyllama/processing/corpus_analysis/vocab_count_{language}.json', 'r') as f:
        data = json.load(f)
        language_dist_dict[language] = data
    print(language, len(data))
```

```python
# print the top 10 words in each language
for language in ['ind', 'vie','thai', 'en']:
    # sort key:value by value
    sorted_pairs = [(int(k), v) for k, v in sorted(language_dist_dict[language].items(), key=lambda item: item[1], reverse=True)]
    sored_decoded_pairs = [(k, (idx2token[k]), v) for k, v in sorted_pairs]
    print("language {} ==============================================================================".format(language))
    print(language, sored_decoded_pairs[:200])
```

```python
total_tokens_count_dict = {k:sum(v.values()) for k,v in language_dist_dict.items()}
# asssert all equal
assert len(set(total_tokens_count_dict.values())) == 1
```

```python
print(total_tokens_count_dict)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency, f_oneway

# Example token frequencies in each language
# token_frequencies = {
#     'token1': [0.3, 0.1, 0.05, 0.55],
#     'token2': [0.2, 0.4, 0.1, 0.3],
#     # Add more tokens as needed
# }

languages=['ind', 'vie','thai', 'en']

# token_frequencies = {
#     i: [language_dist_dict[language].get(str(i),0.01)/total_tokens_count_dict[language] for language in languages] for i in range(len(idx2token))
# }
token_frequencies = {
    i: [language_dist_dict[language].get(str(i),0.01) for language in languages] for i in range(len(idx2token))
}

results = []

for token, frequencies in token_frequencies.items():
    # Convert to contingency table
    counts = np.array(frequencies)  # Scale to large counts for chi-square
    contingency_table = np.array([counts, total_tokens_count_dict['en'] - counts])

    # Perform Chi-Square test
    chi2, p, dof, ex = chi2_contingency(contingency_table)

    # Perform ANOVA
    # f_statistic, p_value_anova = f_oneway(frequencies)


    # Calculate mean and standard deviation
    mean_freq = np.mean(frequencies)
    std_dev = np.std(frequencies)

    # Identify language-specific tokens
    language_specific = np.array(frequencies) > (mean_freq + std_dev)
    dominant_language = languages[np.argmax(frequencies).item()] if np.any(language_specific) else None

    # Append results
    results.append({
        'token': token,
       'chi2_statistic': chi2,
        'chi2_p_value': p,
        #'f_statistic': f_statistic,
        # 'anova_p_value': p_value_anova,
        'language_specific': language_specific,
        'dominant_language': dominant_language,
    'freq': frequencies,
    })

# Convert results to DataFrame for easier viewing
results_df = pd.DataFrame(results)
print(results_df)
```
